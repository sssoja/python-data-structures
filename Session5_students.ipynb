## Part 1: Introduction to Python Classes

### Basic Concepts of OOP:

##### 1. Class:



*   A class is a blueprint for creating objects (a particular data structure), providing initial values for state (member variables or attributes), and implementations of behavior (member functions or methods).
*   You can think of a class as a template for the objects that are based on it.



````python
class Dog:
    pass
````

Here, Dog is a simple class with no attributes or methods.

#### 2. Object:

* An object is an instance of a class. When class is defined, no memory is allocated but when it is instantiated (i.e., an object is created), memory is allocated.
* Objects are the instances of a class created with specific data.

````python
my_dog = Dog()
````

Here, my_dog is an object or instance of the class Dog.

#### 3. The __init__ Method:

* __init__ is a reseved method in Python classes. It is known as a constructor in object oriented concepts. This method is called when an object is created from a class and it allows the class to initialize the attributes of the class.
* The first argument, **self**, represents the instance of the class and allows accessing the attributes and methods of the class.

##### **Example**

````python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
````

Here, name and age are attributes. self.name and self.age are instance variables. The values provided when you create a new Dog instance are stored in these instance variables.

#### 4.Methods

Methods are functions defined inside the body of a class, and they are used to define the behaviors of an object.

##### Example

````python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        print(f"{self.name} says woof!")
````

Here, bark is a method that all Dog instances can perform.

#### 5. The self Parameter:

* **self** represents the instance of the class and allows accessing the attributes and methods of the class. It binds the attributes with the given arguments.
* **self** must be the first parameter of any function in the class.

#### 6. Inheritance

* Inheritance allows us to define a class that inherits all the methods and properties from another class.
* Parent class (or Base class): the class being inherited from.
* Child class (or Derived class): the class that inherits from another class.

##### Example

````python
class Animal:
  def __init__(self, name):  # parent class constructor
    self.name = name

class Dog(Animal):  # inherits from Animal
  def bark(self):
    print(f"{self.name} barks.")
````

````python
grillo = Dog('Grillo')
grillo.bark()
````

Here, Dog inherits from Animal, so every Dog is an Animal.

#### 7. Encapsulation:

Encapsulation is the bundling of data with the methods that operate on that data. It restricts direct access to some of the object's components, which can prevent the accidental modification of data.

#### 8. Polymorphism:

Polymorphism allows the same interface to be used for different underlying forms (data types). This means that different classes can define methods that have the same name, but the methods can behave differently based on which class is being used.

### Defining a Simple Class

* Classes in Python provide a means of bundling data and functionality together.

*   Creating a new class creates a new type of object, allowing new instances of that type to be made.
*   Each class instance can have attributes attached to it for maintaining its state.
* Class instances can also have methods (defined by its class) for modifying its state.


  


````python
# Defining a Simple Class
class Dog:
    # Class Attribute
    species = 'mammal'

    # Initializer / Instance Attributes
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # Instance method
    def description(self):
        return f"{self.name} is {self.age} years old"

    # Another instance method
    def speak(self, sound):
        return f"{self.name} says {sound}"

# Instantiate the Dog class
buster = Dog("Buster", 6)

# Access the instance attributes
print(f"{buster.name} is a {buster.species} and {buster.age} years old.")
print(buster.description())
print(buster.speak("Guau guau"))
````

### Creating a Base Class and Subclass

````python
# Base class
class Bird:
    def __init__(self):
        print("Bird is ready")

    def whoisThis(self):
        print("Bird")

    def swim(self):
        print("Swim faster")

# Derived class
class Penguin(Bird):
    def __init__(self):
        super().__init__()  # call super() function
        print("Penguin is ready")

    def whoisThis(self):
      super().whoisThis()
      print("Penguin")

    def run(self):
        print("Run faster")

peggy = Penguin()
peggy.whoisThis()
peggy.swim()
peggy.run()
````

## Part 2: Introduction to Matplotlib

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy.

### Basic Plotting

````python
# Basic Plotting with Matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, '-b', label='Sine wave')
plt.title('Simple Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='upper right')
plt.show()
````

###  Scatter Plots

Scatter plots are used to plot data points on a horizontal and vertical axis in the attempt to show how much one variable is affected by another. Each row in the data table is represented by a dot or marker whose position depends on its values in the columns set on the X and Y axes.


````python
# Creating a scatter plot
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(6, 4))
plt.scatter(x, y)
plt.title('Random Scatter Plot')
plt.xlabel('Random x values')
plt.ylabel('Random y values')
plt.show()
````

### Box Plots

Box plots are a standard way of displaying the distribution of data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. They are particularly useful for highlighting outliers and understanding the spread of the data.


````python
# Creating a box plot
data = np.random.randn(100)

plt.figure()
plt.boxplot(data)
plt.title('Box Plot of Random Data')
plt.ylabel('Values')
plt.show()
````

````python
# Exercise: Create box plots for different sets of data to compare their distributions.
data1 = np.random.normal(0, scale=0.2, size=100)
data2 = np.random.normal(0, scale=1.0, size=100)
data3 = np.random.normal(0, scale=2.0, size=100)

plt.figure()
plt.boxplot([data1, data2, data3], labels=['Low Variance', 'Medium Variance', 'High Variance'])
plt.title('Comparing Distributions')
plt.ylabel('Values')
plt.show()
````
